```python
df1=df.drop(columns=['Photo','Flag','Club Logo','Preferred Foot','Weak Foot',
                 'Skill Moves','Work Rate','Body Type','Jersey Number','Crossing','Finishing','HeadingAccuracy','ShortPassing','Volleys','Dribbling','Curve','FKAccuracy',
                'BallControl','Acceleration','SprintSpeed','Agility','Balance','ShotPower','Jumping','Stamina','LongShots','Aggression',
                'Interceptions','Positioning','Vision','Composure','Marking','StandingTackle','GKDiving','SlidingTackle',
                'GKKicking','GKPositioning','GKReflexes','Loaned From','LongPassing','Reactions','Strength','Penalties','GKHandling'],axis=1)
```

```python
df1.dropna(subset=['Club'],inplace=True)
df1=df1.dropna(subset=['Joined'])
df1[df1['Joined'].isnull()]
```

```python
list1=[]
for i in df1['Value']:
    i=i.replace("€","")
    if 'M' in i:
        i=i.replace("M","")
        i=float(i)*1000000
    elif 'K' in i:
        i=i.replace("K","")
        i=float(i)*1000
    list1.append(i)
list1    
```

```python
list2=[]
for i in df1['Release Clause']:
    i=str(i).replace("€","")
    if 'M' in i:
        i=str(i).replace("M","")
        i=float(i)*1000000
    elif 'K' in i:
        i=str(i).replace("K","")
        i=float(i)*1000
    list2.append(i)
list2    
```

```python
list3=[]
for i in df1['Wage']:
    i=str(i).replace("€","")
    if 'M' in i:
        i=str(i).replace("M","")
        i=float(i)*1000000
    elif 'K' in i:
        i=str(i).replace("K","")
        i=float(i)*1000
    list3.append(i)
list3 
```

```python
df1['Value']=list1
df1['Release Clause']=list2
df1['Wage']=list3
df1['Value']=df1['Value'].astype('float')
df1['Wage']=df1['Wage'].astype('float')
df1['Release Clause']=df1['Release Clause'].astype('float')
df1.info()
```

```python
df1['Release Clause']=df1['Release Clause'].fillna(df1['Release Clause'].mean())
df1['Release Clause'].isnull().sum()
```

```python
df1
```

```python
df1['Joined']=pd.DatetimeIndex(df1['Joined']).year
df1
```

```python
df1=df1.dropna(subset=['Joined'])
df1['Joined'].isnull().sum()
```

```python
df1['Joined']=df1['Joined'].astype('int')
df1['Joined']
```

```python
df1['Contract Valid Until']=pd.to_datetime(df1['Contract Valid Until'])
df1
```

```python
list5=[]
for i in df1['Height']:
    i=i.replace("'",".")
    list5.append(float(i))
df1['Height']=list5    
df1
```

```python
list6=[]
for i in df1['Weight']:
    i=i.replace('lbs','')
    list6.append(float(i))
df1['Weight']=list6
df1
```
